```python
import pdfplumber
import re

def extract_director_info(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ""
        for page in pdf.pages:
            text += page.extract_text()

    # Define pattern for DIN (Director Identification Number)
    din_pattern = r"\b\d{8}\b"

    # Define pattern for director type (independent/executive)
    director_type_pattern = r"(independent|executive)"

    # Define pattern for director name
    name_pattern = r"Mr\.?\s[A-Z][a-z]+\s[A-Z][a-z]+"

    # Extract director name
    director_name_match = re.search(name_pattern, text)
    director_name = director_name_match.group() if director_name_match else None

    # Extract director type
    director_type_match = re.search(director_type_pattern, text, re.IGNORECASE)
    director_type = director_type_match.group(0) if director_type_match else None

    # Extract DIN
    din_match = re.search(din_pattern, text)
    DIN = din_match.group(0) if din_match else None

    return director_name, director_type, DIN

# Example usage
pdf_path = "PolyPlex.pdf"
director_name, director_type, DIN = extract_director_info(pdf_path)
print("Director Name:", director_name)
print("Director Type:", director_type)
print("DIN:", DIN)
```

```python
import pdfplumber
import re

def extract_director_info(pdf_path):
    director_info = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()

            # Define pattern for DIN (Director Identification Number)
            din_pattern = r"DIN:\s*(\d{8})"

            # Define pattern for director type (independent/executive)
            director_type_pattern = r"(?i)(independent|executive)\s*Director"

            # Define pattern for director name
            name_pattern = r"Mr\.?\s+([A-Z][a-z]+\s+[A-Z][a-z]+)"

            # Extract director names, types, and DINs
            names = re.findall(name_pattern, text)
            types = re.findall(director_type_pattern, text)
            DINs = re.findall(din_pattern, text)

            # Associate each director's information together
            for name, director_type, DIN in zip(names, types, DINs):
                director_info.append({"Name": name.strip(), "Type": director_type.strip().lower(), "DIN": DIN.strip()})

    return director_info

# Example usage
pdf_path = "Vinati.pdf"
director_info = extract_director_info(pdf_path)

for director in director_info:
    print("Name:", director["Name"])
    print("Type:", director["Type"])
    print("DIN:", director["DIN"])
    print()
```

```python
import streamlit as st
import PyPDF2 
# Or Camelot
import spacy
# Load a pre-trained NER model (spaCy or NLTK)
nlp = spacy.load("en_core_web_sm")  # Replace with your chosen model

def extract_pdf_text(pdf_file):
  # Implement text extraction using PyPDF2 or Camelot
  pdf_reader = PyPDF2.PdfReader(pdf_file)
  text = ""
  for page in pdf_reader.pages:
    text += page.extract_text()
  return text

def extract_director_info(text, nlp):
  # Process text with your NER model to extract information
  doc = nlp(text)
  extracted_data = {}
  for entity in doc.ents:
    if entity.label_ in ["NAME", "DIRECTOR_TYPE", "DIN"]:  # Adjust labels based on your model
      extracted_data[entity.label_] = entity.text_
  return extracted_data

def main():
  st.title("Director Information Extraction")
  uploaded_file = st.file_uploader("Upload PDF")
  if uploaded_file is not None:
    # Extract text from uploaded PDF
    pdf_text = extract_pdf_text(uploaded_file.read())
    
    # Extract information using NER model
    extracted_data = extract_director_info(pdf_text, nlp)
    
    # Display extracted information
    if extracted_data:
      st.write("Extracted Information:")
      for key, value in extracted_data.items():
        st.write(f"{key}: {value}")
    else:
      st.write("No relevant information found in the PDF.")

if __name__ == "__main__":
  main()
```
